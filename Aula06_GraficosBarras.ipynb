# **Criação** de gráficos


**Dados Inseridos**

````python
import pandas as pd # Biblioteca de dados (filtro e etc..)
import matplotlib.pyplot as plt # Bliblioteca de graficos

# Dados
dados = pd.DataFrame({ #dentro dos colchetes utilza os : no lugar do =
    "Frutas" : [ "Maça", "Banana", "Laranja", "Manga" ], #Virgula para separar
    "Vendas" : [ 50, 30, 70, 40 ]
})

#Grafico / bar = grafico de barras
# plt.bar( x , y )
plt.bar(dados["Frutas"], dados["Vendas"], color="darkblue")

#Legendas
plt.title("Vendas de Frutas(un)", color="darkblue")
plt.xlabel("Frutas", color="darkblue")
plt.ylabel("Quantidade", color="darkblue")
plt.grid(axis="y", linestyle="--", alpha=0.7, color="grey")
````

### **Vendas por Faixa de Preço**

**Tarefa:**

Crie faixas de preço:

- Até R$ 100

- De R 101 a R$ 200

- De R 201 a R$ 300

- Acima de R$ 300

E exiba um gráfico de barras mostrando o total de **quantidade de compras** em cada faixa.

**Dica:**

```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

# 1. Upload e leitura do arquivo
uploaded = files.upload()
nome_arquivo = next(iter(uploaded))
df = pd.read_excel(nome_arquivo)

# 2. Criar as faixas de preço (bins)
bins = [0, 100, 200, 300, float('inf')]  # limites das faixas
labels = ['Até R$100', 'R$101–R$200', 'R$201–R$300', 'Acima de R$300']

# 3. Criar uma nova coluna com a faixa correspondente
df['Faixa de Preço'] = pd.cut(df['Preço (R$)'], bins=bins, labels=labels, right=True)

print(df)

# 4. Agrupar por faixa e somar a quantidade de compras
vendas_por_faixa = df.groupby('Faixa de Preço')['Quantidade de Compras'].sum()

# 5. Criar o gráfico de barras
plt.bar(vendas_por_faixa.index, vendas_por_faixa.values, color='royalblue')

# 6. Personalização
plt.title("Vendas por Faixa de Preço")
plt.xlabel("Faixa de Preço")
plt.ylabel("Quantidade Total de Compras")
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

### **Receita Média por Tipo de Jogo**

**Tarefa:**

Agrupe os jogos pelo tipo e calcule a **receita média por jogo** (Receita Total ÷ Quantidade de Jogos por Tipo).

Exiba em um gráfico de barras.

Dica:

```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

# 1. Upload e leitura do arquivo
uploaded = files.upload()
nome_arquivo = next(iter(uploaded))
df = pd.read_excel(nome_arquivo)

# 2. Agrupar por tipo de jogo e calcular receita total e contagem
agrupado = df.groupby("Tipo de Jogo").agg({
    "Preço (R$)": "sum",           # Soma total da receita por tipo
    "Nome do Jogo": "count"        # Quantidade de jogos em cada tipo
}).rename(columns={"Preço (R$)": "Receita Total", "Nome do Jogo": "Qtd Jogos"})

# 3. Calcular receita média por jogo
agrupado["Receita Média"] = agrupado["Receita Total"] / agrupado["Qtd Jogos"]

# 4. Ordenar do maior para o menor
agrupado = agrupado.sort_values(by="Receita Média", ascending=False)

# 5. Gráfico de barras
plt.figure(figsize=(8,5))
plt.bar(agrupado.index, agrupado["Receita Média"], color='mediumseagreen')

# 6. Personalização
plt.title("Receita Média por Categoria de Jogo")
plt.xlabel("Tipo de Jogo")
plt.ylabel("Receita Média (R$)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')

# 7. Mostrar valores sobre as barras
for i, v in enumerate(agrupado["Receita Média"]):
    plt.text(i, v + (v*0.02 if v != 0 else 1), f"R$ {v:.2f}", ha='center', va='bottom', fontsize=9)

plt.show()
```
